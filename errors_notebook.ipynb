```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sympy import *
from storage import get_data_frame
```

```python
def estetica(figura,x_name = '',y_name = '',title = '',w = 900,h = 450):
    figura.update_xaxes(showgrid=True,showline=True, linewidth=2, linecolor='black', mirror=True,title = x_name)#'cccccccccccccccc')
    figura.update_yaxes(showgrid=True,showline=True, linewidth=2, linecolor='black', mirror=True,title = y_name)


    figura.update_layout(
        margin=dict(l=20, r=20, t=40, b=20),
        plot_bgcolor="white",
        template="plotly_white",
        title=title,
        autosize=False,
        width=w,
        height=h,
        #yaxis = dict(range=[0,1]),
        #xaxis = dict(range=[0,1]),
        title_font = dict(size = 20),
        legend=dict(orientation = 'v',
                    font = dict(size = 18),
                     yanchor="top",
                     y=0.99,
                     xanchor="right",
                     x=1,
                     bgcolor='rgba(255,255,255,0.8)',
                   ),
        font = dict(size=20)
    )
```

```python
def show_with_estetica(x_data, y_data, x_name = '', y_name = '', title = '', w = 900, h = 450):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x_data, y=y_data, mode='lines'))
    estetica(fig, x_name, y_name, title, w, h)
    fig.show()
```

```python
def show_multiple_with_estetica(x_data, y_named_datas, x_name = '', y_name = '', title = '', w = 900, h = 450):
    fig = go.Figure()
    for y_named_data in y_named_datas:
        y_data, trace_name = y_named_data
        fig.add_trace(go.Scatter(x=x_data, y=y_data, mode='lines', name = trace_name))
    estetica(fig, x_name, y_name, title, w, h)
    fig.show()
```

```python
def to1Digit(x):
    return float('%.1g' % x)
```

```python
df = get_data_frame("video2")
```

```python
show_with_estetica(df['time'], df['theta_wrist'], x_name = 'time (s)', y_name = 'theta_wrist (rad)', title = 'Wrist angle vs. Time')
```

```python
T_values = [3.2, 3.4, 2.8, 3]
T_probable = np.mean(T_values)
T_std = np.std(T_values)
Delta_T = np.sqrt(T_std ** 2 + (1 / 30) ** 2)
print(f'T = [{T_probable}  {to1Digit(Delta_T)}]')
```

```python
T = Symbol('T')
f = 1 / T
f_T = f.diff(T)
print(f'Derivada de f respecto de T -> {f_T}')
```

```python
f_probable = 1 / T_probable
diff_f_T = -1 / T_probable ** 2
Delta_f = np.sqrt(diff_f_T ** 2 * Delta_T ** 2)
print(f'f = [{f_probable}  {to1Digit(Delta_f)}]')
```

```python
f = Symbol('f')
w = 2 * np.pi * f
w_f = w.diff(f)
print(f'Derivada de w respecto de f -> {w_f}')
```

```python
w_probable = 2 * np.pi * f_probable
diff_w_f = 2 * np.pi
Delta_w = np.sqrt(diff_w_f ** 2 * Delta_f ** 2)
print(f'w = [{w_probable}  {to1Digit(Delta_w)}]')
```

```python
A_values = [1.14, 1.1, 1.14, 1.1, 1.05]
A_probable = np.mean(A_values)
A_std = np.std(A_values)
Delta_A = np.sqrt(A_std ** 2)
print(f'A = [{A_probable}  {to1Digit(Delta_A)}]')
```

```python
C_values = [1.93, 2.01, 2.08, 2.03, 2.01]
C_probable = np.mean(C_values)
C_std = np.std(C_values)
Delta_C = np.sqrt(C_std ** 2)
print(f'C = [{C_probable}  {to1Digit(Delta_C)}]')
```

```python
g_probable = A_probable * np.cos(w_probable * df['time']) + C_probable
```

```python
show_multiple_with_estetica(df['time'], [(df['theta_wrist'], 'theta_wrist'), (g_probable, 'Acos(wt)+C')], x_name = 'time (s)', y_name = 'angle (rad)', title = 'Wrist angle vs. Time')
```
