```python
import pandas as pd
import plotly.graph_objects as go
from storage import get_data_frame
```

```python
def estetica(figura,x_name = '',y_name = '',title = '',w = 900,h = 450):
    figura.update_xaxes(showgrid=True,showline=True, linewidth=2, linecolor='black', mirror=True,title = x_name)#'cccccccccccccccc')
    figura.update_yaxes(showgrid=True,showline=True, linewidth=2, linecolor='black', mirror=True,title =y_name)


    figura.update_layout(
        margin=dict(l=20, r=20, t=40, b=20),
        plot_bgcolor="white",
        template="plotly_white",
        title=title,
        autosize=False,
        width=w,
        height=h,
        #yaxis = dict(range=[0,1]),
        #xaxis = dict(range=[0,1]),
        title_font = dict(size =20),
        legend=dict(orientation = 'v',
                    font = dict(size = 18),
                     yanchor="top",
                     y=0.95,
                     xanchor="right",
                     x=1,
                     bgcolor='rgba(255,255,255,0)',
                   )
    )
```

```python
def show_with_estetica(x_data, y_data, x_name = '', y_name = '', title = '', w = 900, h = 450):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x_data, y=y_data, mode='lines'))
    estetica(fig, x_name, y_name, title, w, h)
    fig.show()
```

```python
def show_multiple_with_estetica(x_data, y_named_datas, x_name = '', y_name = '', title = '', w = 900, h = 450):
    fig = go.Figure()
    for y_named_data in y_named_datas:
        y_data, y_name = y_named_data
        fig.add_trace(go.Scatter(x=x_data, y=y_data, mode='lines', name = y_name))
    estetica(fig, x_name, y_name, title, w, h)
    fig.show()
```

```python
df = get_data_frame("video2")
```

```python
show_with_estetica(df['time'], df['rx_wrist'], x_name = 'time (s)', y_name = 'rx_wrist (m)', title = 'Wrist position in X vs. Time')
```

```python
show_with_estetica(df['time'], df['ry_wrist'], x_name = 'time (s)', y_name = 'ry_wrist (m)', title = 'Wrist position in Y vs. Time')
```

```python
show_with_estetica(df['time'], df['vy_wrist'], x_name = 'time (s)', y_name = 'vy_wrist (m/s)', title = 'Wrist velocity in Y vs. Time')
```

```python
show_with_estetica(df['time'], df['ay_wrist'], x_name = 'time (s)', y_name = 'ay_wrist (m/s^2)', title = 'Wrist acceleration in Y vs. Time')
```

```python
show_with_estetica(df['time'], df['force_bicep'], x_name = 'time (s)', y_name = 'force_bicep (N)', title = 'Bicep force vs. Time')
```

```python
show_multiple_with_estetica(df['time'], [(df['work_i'], 'work from torque'), (df['work_i_from_energy'], 'work from energy')], x_name = 'time (s)', y_name = 'work (J)', title = 'Work vs. Time')
```
